# Download Pokemon Images with Selenium & Beautifulsoup4 & Requests

```python
# %pip install beautifulsoup4
# %pip install requests
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from time import sleep
```

```python
# https//zukan.pokemon.co.jp のポケモン個別url
base_url = "https://zukan.pokemon.co.jp"
urls = []
for i in range(898):
    index = i+1
    if index//10 == 0:
        index_str = f"00{index}"
    elif index/10 < 10:
        index_str = f"0{index}"
    else:
        index_str = str(index)
    urls.append(f"{base_url}/detail/{index_str}")
```

```python
urls[0] #フシギダネ(No.001)
```

```python
# selenium高速化
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--single-process')
options.add_argument('--disable-application-cache')
options.add_argument('--ignore-certificate-errors')
options.add_argument('--start-maximized')
driver = webdriver.Chrome(options=options)
```

```python
# 個別ページから画像を取得
def get_images(start=0,end=898):
    for i in range(start,end):
        driver.get(urls[i])
        sleep(0.1) # flexの読み込み待ち
        soup = BeautifulSoup(driver.page_source,'html.parser')
        header = soup.find_all('div',class_="inner__header__detail")
        img_tag = header[0].find_all(class_='img')
        img_url = img_tag[0].get('src')
        print(img_url)
        response = requests.get(img_url)
        with open(f"Pokemon_images/img{urls[i].split('/')[-1]}.png", 'wb') as f:
            f.write(response.content)
```

```python
get_images(start=0,end=898)
```
